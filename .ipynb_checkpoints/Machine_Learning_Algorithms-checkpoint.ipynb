Ridge Regression Algorithm Implementation

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge_regressor = Ridge()
```

```python
ridge_regressor
```

```python
parameteres = {'alpha': [1, 2, 5, 10, 20, 30, 40, 60, 70, 80, 90]}
ridge_cv = GridSearchCV(ridge_regressor, parameteres, scoring='neg_mean_squared_error', cv = 5)
ridge_cv.fit(X_train, y_train)
```

```python
print(ridge_cv.best_params_)
```

```python
print(ridge_cv.best_score_)
```

```python
# predict
ridge_predict = ridge_cv.predict(X_test)
```

```python
ridge_predict
```

```python
import seaborn as sns
sns.displot(ridge_predict -  y_test, kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ridge_predict, y_test)
score
```
