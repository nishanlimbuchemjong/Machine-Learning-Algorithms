Logistic Regression Implementation

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# getting the unique species
df['species'].unique()
```

```python
# checking wherther there is null value or not
df.isnull().sum()
```

```python
# for Binary classificaton the output should be of two number. 
# But the given dataset have three output i.e. ['setosa', 'versicolor', 'virginica'] which we can say that the given dataset is multiclass classification
# So we will do a Bianry classification. To do Binary classificatioh, remove one output to make it two numbers of output
# Removing one output to make it two number of output:
df = df[df['species'] != 'setosa']  # display all the output that does not contains species='setosa'
```

```python
df
```

```python
# we can't give the species name to the model. so we need to convert it into numberal form
df['species'] = df['species'].map({'versicolor':0, 'virginica':1})
```

```python
df['species']
```

```python
df.head()
```

```python
# splitting the dataset into indepenet and dependent features
```

```python
# In the given dataset, sepal_length, speal_width, petal_length, and petal_width are independent features,
# So we need to put all these into one variable
X = df.iloc[:, :-1]  # taking all the columns except the last one i.e. species which are independent
y = df.iloc[:, -1]   # taking the last columns i.e, species which is dependent 
```

```python
X
```

```python
y
```

```python
# Use train test split
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)

# test data are used for validation where model predict and gives the result
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
```

```python
# How all parameters are used for hyper parameter tunning, so we use GridSearchCV
from sklearn.model_selection import GridSearchCV

parameter = {'penalty':['l1','l2', 'elasticNet'], 'C':[1, 2, 3, 4, 5, 6, 8, 10, 20, 30, 40, 50], 'max_iter':[100, 200, 300]}
```

```python
# classifier_regressor is a name of model
classifier_regressor = GridSearchCV(classifier, param_grid=parameter, scoring='accuracy', cv=5)
```

```python
classifier_regressor.fit(X_train, y_train)
```

```python
# printing best params
print(classifier_regressor.best_params_)
```

```python
# printing accuracy score
print(classifier_regressor.best_score_)
```

```python
# prediction
y_predict = classifier_regressor.predict(X_test)
y_predict
```

```python
# accuracy score
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_predict, y_test)
score
```

```python
print(classification_report(y_predict, y_test))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# we cna also import in another form that gives a well dataframe
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# taking independent features and dependent features
X = df.iloc[:, :-1]
y = data.target
```

```python
X
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
# importing decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
# post prunning 
treemodel = DecisionTreeClassifier(max_depth=2)
```

```python
treemodel.fit(X_train, y_train)
```

```python
# for contructing tree
from sklearn import tree

plt.figure(figsize=(15, 10))
tree.plot_tree(treemodel, filled=True)
```

```python
# prediction
y_predict = treemodel.predict(X_test)
y_predict
```

```python
# checking accuracy
from sklearn.metrics import accuracy_score, classification_report
```

```python
# for accuracy
score = accuracy_score(y_predict, y_test)
score
```

```python
# for classification report
print(classification_report(y_predict, y_test))
```
